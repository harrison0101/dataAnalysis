# Limpeza e padronização

```python
#Limpeza e padronização

for col in ['crime_type', 'city', 'state', 'location_description', 'victim_gender', 'victim_race']:
	df[col] = df[col].astype(str).str.strip()
```

```python
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], errors='coerce')
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day']=df['datetime'].dt.day
df['hour']=df['datetime'].dt.hour
df['day_of_week']=df['datetime'].dt.dayofweek
df['weekday_name']=df['datetime'].dt.day_name()
```

```python
df=df[(df['victim_age'] >= 0) & (df['victim_age'] <= 120)]
```

```python
bins = [0,12,17,24,34,44,54,64,120]
labels = ['0-12','13-17','18-24','25-34','35-44','45-54','55-64','65+']
df['age_group'] = pd.cut(df['victim_age'], bins=bins, labels=labels, include_lowest=True, right=True)
```

```python
gender_map = {'m':'Male','M':'Male','male':'Male',
              'f':'Female','F':'Female','female':'Female'}
df['victim_gender'] = df['victim_gender'].replace(gender_map)
```

```python
df = df.drop_duplicates(subset=['id'])
```

```python
assert df['datetime'].isna().sum() == 0, "Há linhas sem datetime válido."
assert df['id'].is_unique, "IDs duplicados encontrados."
```

```python
df.to_csv('/content/crime_safety_dataset_clean.csv', index=False)
print("Arquivo salvo em /content/crime_safety_dataset_clean.csv")
```

# Gráficos

```python
df['crime_type'].value_counts().head(10).plot(kind='bar'); plt.title('Top 10 Tipos de Crime'); plt.xlabel('Tipo'); plt.ylabel('Ocorrências'); plt.xticks(rotation=45); plt.show()
```

```python
df.groupby('hour').size().reindex(range(24), fill_value=0).plot(kind='line', marker='o'); plt.title('Ocorrências por Hora'); plt.xlabel('Hora'); plt.ylabel('Ocorrências'); plt.grid(True, axis='y'); plt.show()
```

```python
df['city'].value_counts().head(10).plot(kind='bar'); plt.title('Top 10 Cidades'); plt.xlabel('Cidade'); plt.ylabel('Ocorrências'); plt.xticks(rotation=45); plt.show()
```

```python
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
df['weekday_name'].value_counts().reindex(order).plot(kind='bar'); plt.title('Ocorrências por Dia da Semana'); plt.xlabel('Dia'); plt.ylabel('Ocorrências'); plt.xticks(rotation=45); plt.show()
```
